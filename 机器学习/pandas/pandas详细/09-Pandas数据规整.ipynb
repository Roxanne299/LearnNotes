# 数据合并

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right)
```

```python
pd.merge(left,right,on='key')
```

```python
#处理重复列名
df_obj1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
'data' : np.random.randint(0,10,7)})
df_obj2 = pd.DataFrame({'key': ['a', 'b', 'd'],
'data' : np.random.randint(0,10,3)})
```

```python
df_obj1
```

```python
df_obj2
```

```python
print(pd.merge(df_obj1, df_obj2, on='key', suffixes=('_left', '_right')))
```

```python
#按索引连接
df_obj1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
'data1' : np.random.randint(0,10,7)})
df_obj2 = pd.DataFrame({'data2' : np.random.randint(0,10,3)}, index=['a', 'b', 'd'])
```

```python
df_obj1 
```

```python
df_obj2
```

```python
print(pd.merge(df_obj1, df_obj2, left_on='key', right_index=True))  #rightindex表示将右侧的index作为连接的键
```

```python
print(pd.merge(df_obj1, df_obj2, left_on='key', right_index=True))
```

## join

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=['a', 'c', 'e'],
                     columns=['语文', '数学'])

right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=['b', 'c', 'd', 'e'],
                      columns=['英语', '综合'])
```

```python
left2
```

```python
right2
```

```python
pd.merge(left2,right2,how='outer',left_index=True,right_index=True)
```

```python
# 按照索引合并  但是要求没有重叠的列
left2.join(right2,how='outer')
```

# 重塑和轴向旋转

### 重塑层次化索引

```python
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
                    index=pd.Index(['老王', '小刘'], name='姓名'),
                    columns=pd.Index(['语文', '数学', '英语'],name='科目'))
data
```

```python
r = data.stack()  #
r
```

```python
type(r)
```

```python
r.unstack()
```

```python
r.unstack(0)
```

```python
r.unstack('姓名')
```

```python
a1 = pd.Series(np.arange(4),index=list('abcd'))
a1
```

```python
a2 = pd.Series([4,5,6],index=list('cde'))
a2
```

```python
s1 = pd.concat([a1,a2],keys=['data1','data2'])
s1
```

```python
type(s1)
```

```python
s1.unstack()
```

```python
# stack()默认过滤缺失数据
s1.unstack().stack()
```

```python
s1.unstack().stack(dropna=False)
```

### 轴向旋转

```python
df3 = pd.DataFrame({'date':['2018-11-22','2018-11-22','2018-11-23','2018-11-23','2018-11-24'],
                  'class':['a','b','b','c','c'],
                  'values':[5,3,2,6,1]},columns=['date','class','values'])
df3
```

```python
df3.pivot('date','class','values')  #分别是行索引，列索引，填充值
```

```python
df3.pivot('date','class')
```

```python
df3.set_index(['date','class']).unstack('class')
```
