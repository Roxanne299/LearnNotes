## JSON数据

```python
import json
```

```python
obj = """
{"name": "Wes",
 "places_lived": ["United States", "Spain", "Germany"],
 "pet": null,
 "siblings": [{"name": "Scott", "age": 30, "pets": ["Zeus", "Zuko"]},
              {"name": "Katie", "age": 38,
               "pets": ["Sixes", "Stache", "Cisco"]}]
}
"""
```

```python
# 将json字符串转换为python形式
res = json.loads(obj)
res
```

```python
# 将python对象转换为json格式
res1 = json.dumps(res)
res1
```

```python
sib = pd.DataFrame(res['siblings'],columns=['name','age'])
sib
```

## HDF5格式

```python
frame = pd.DataFrame({'a': np.random.randn(100)})
frame
```

```python
store = pd.HDFStore('mydata.h5')
```

```python
store['obj1'] = frame
```

```python
store['obj1_col'] = frame['a']
```

```python
store
```

```python
store['obj1']
```

```python
#两种存储模式 fixed table
store.put('obj2',frame,format='table') ##== store['obj2'] = frame
```

```python
store.select('obj2',where =['index>=10 and index <= 15'])
```

# 3.Web APls交互

```python
import requests
url='https://api.github.com/repos/pandas-dev/pandas/issues'
resp = requests.get(url)
resp
```

```python
#响应对象的json方法返回一个包含解析过的json字典，加载到一个python对象中
data = resp.json()
```

```python
data[0]['title']
```

```python
issues = pd.DataFrame(data,columns=['number','title','labels','state'])
issues
```

# 4.数据库交互

pandas
sqlalchemy
pymysql

```python
# 导入必要模块
import pandas as pd
from sqlalchemy import create_engine

#初始化数据库连接
#用户名root 密码   端口 3306  数据库 db2
engine = create_engine('mysql+pymysql://root:@localhost:3306/db2')
#查询语句
sql = '''
    select * from class;
'''
#两个参数   sql语句  数据库连接
df = pd.read_sql(sql,engine)
df
```

```python
#新建
df = pd.DataFrame({'id':[1,2,3,4],'num':[34,56,78,90]})
df = pd.read_csv('ex1.csv')
# #写入到数据库
df.to_sql('df2',engine,index=False)
print("ok")
```
