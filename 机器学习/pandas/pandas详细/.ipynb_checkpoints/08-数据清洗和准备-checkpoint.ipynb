### 利用函数或映射进行数据转换

```python
data = pd.DataFrame({'food': ['Apple', 'banana', 'orange','apple','Mango', 'tomato'],
                    'price': [4, 3, 3.5, 6, 12,3]})
data
```

```python
meat = {'apple':'fruit',
       'banana':'fruit',
       'orange':'fruit',
       'mango':'fruit',
       'tomato':'vagetables'}
```

```python
#值小写
low = data['food'].str.lower()
low
```

```python
data['class'] = low.map(meat)
data
```

```python
data['class1'] = data['food'].map(lambda x:meat[x.lower()])
data
```

```python
data['class1'] = data['food'].map(lambda x: meat[x.lower()])
data
```

### 离散化和面元划分      ----  分阶段

0-1e9   1-1e5
6 8 9 4  ===> 2 3 4 1

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

```python
# 面元 bin
bins = [18,25,35,60,100]
```

```python
cats = pd.cut(ages,bins)
cats
```

```python
cats.codes
```

```python
cats.categories
```

```python
pd.value_counts(cats)
```

```python
pd.cut(ages,[18,26,36,61,100],right=False)
```

```python
names = ['青年','年轻人','中年','老年']
```

```python
pd.cut(ages,bins,labels=names)
```

```python
data = np.random.rand(20)
data
```

```python
pd.cut(data,4,precision = 2)   #小数位数2
```

```python
# qcut函数
data = np.random.randn(1000)
data
```

```python
cats = pd.qcut(data,4)
cats
```

```python
pd.value_counts(cats)
```

```python
cats = pd.qcut(data,[0,0.1,0.5,0.9,1.])
```

```python
pd.value_counts(cats)
```

### 正则表达式

```python
import re
```

```python
text = "foo  bar\t bat  \tqq"
```

```python
re.split('\s+',text)
```

```python
res = re.compile('\s+')
res.split(text)
```

```python
reg = re.split(res,text)
reg
```

```python
res.findall(text)
```

```python
re.findall(res,text)
```

```python
re.sub(res,'9',text)
```

```python
t1 = re.match('f',text)
print(t1.group())
```

```python
print(re.search('b',text).group())
```
