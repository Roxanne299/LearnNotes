```python
import numpy as np
import pandas as pd
```

```python
guazi_df = pd.read_csv("guazi.csv")
```

```python
guazi_df.head()
```

```python
# 1. 重命名列名
guazi_df.columns=['Model','Years','Mileage','City','Selling_price','Origin_price']
guazi_df.columns
```

```python
# 2. 删除异常值，那些重复的值
guazi_df = guazi_df[guazi_df['Model'] != 'leixing']
guazi_df
```

```python
# 3. 删除重复的值
guazi_df.drop_duplicates(inplace=True)
guazi_df
```

```python
# 4. 处理缺失值，把有缺失值的数据删掉
guazi_df.dropna(axis=0,how="any",inplace=True)
guazi_df
```

```python
# 5. 将年份，里程，售价，原价转换成数值类型
guazi_df['Years'] = guazi_df['Years'].apply(lambda value:np.float(value.replace("年","")))
guazi_df['Mileage'] = guazi_df['Mileage'].apply(lambda value:np.float(value.replace("公里","").replace("万","")))
guazi_df['Selling_price'] = guazi_df['Selling_price'].apply(lambda value:np.float(value.replace("万","").replace("补贴后","")))
guazi_df['Origin_price'] = guazi_df['Origin_price'].apply(lambda value:np.float(value.replace("万","")))
guazi_df.head()
```
