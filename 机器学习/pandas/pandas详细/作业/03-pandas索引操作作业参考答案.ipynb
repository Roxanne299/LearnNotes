```python
import numpy as np
import pandas as pd
```

```python
lianjia_df = pd.read_csv("lianjia.csv")
```

```python
lianjia_df.head()
```

```python
# 1. 观察结构，调整列索引顺序
lianjia_df[["Region","Garden","Layout","Floor","Year","Size","Elevator","Direction","Renovation","Price"]].head()
```

```python
# 2. 增加一个列关于目前状况（state），是否卖出状态随机
import random
states = [0,1]
column_count = lianjia_df.shape[0]
lianjia_df['state'] = [random.choice(states) for x in range(column_count)]
lianjia_df.head()
```

```python
# 3. 删除id这一列无用的数据
# 备注：因为文件中的id不在了，所以先添加一个再删除一下
column_count = lianjia_df.shape[0]
# 先添加
lianjia_df['Id'] = [x for x in range(1,column_count+1)]
# 再删除
lianjia_df.drop(columns=["Id"],inplace=True)
lianjia_df.head()
```

```python
# 4. 查找楼层小于7的房子（这里提取低楼层）
def extract_low_floor(floors):
    low_floors = []
    for floor in floors:
        if "低楼层" in floor:
            low_floors.append(True)
        else:
            low_floors.append(False)
    return low_floors
low_floor_lianjia_df = lianjia_df[extract_low_floor(lianjia_df['Floor'])]
low_floor_lianjia_df.head()
```

```python
low_floor_line_count = low_floor_lianjia_df.shape[0]
low_floor_lianjia_df.loc[0:,'Elevator'] = '无电梯'
low_floor_lianjia_df
```
