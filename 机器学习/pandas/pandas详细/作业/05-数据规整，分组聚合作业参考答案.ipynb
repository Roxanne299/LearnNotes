```python
import numpy as np
import pandas as pd
```

```python
guazi_bj = pd.read_csv("guazi_bj.csv")
guazi_gz = pd.read_csv("guazi_gz.csv")
guazi_sh = pd.read_csv("guazi_sh.csv")
guazi_sz = pd.read_csv("guazi_sz.csv")
```

```python
# 1. 拼接多个csv文件
guazi_df = pd.concat([guazi_bj,guazi_gz,guazi_sh,guazi_sz],axis=0)
guazi_df
```

```python
# 2. 去除重复数据，重新索引
guazi_df.drop_duplicates(inplace=True)
guazi_df.reset_index()
```

```python
# 3. 汽车分自动挡和手动挡，分别多少辆
auto_df = guazi_df[guazi_df['speedbox']=='自动']
manual_df = guazi_df[guazi_df['speedbox']=='手动']
print('自动挡：',len(auto_df.index))
print("手动挡：",len(manual_df.index))
```

```python
# 4. 哪个城市二手车数量最多
guazi_df.groupby("city").count().sort_values("title",ascending=False).head()
```

```python
# 5. 找出深圳哪个品牌的二手车最多
guazi_df[guazi_df['city']=='深圳'].groupby("brand").count().sort_values("title",ascending=False).head()
```
