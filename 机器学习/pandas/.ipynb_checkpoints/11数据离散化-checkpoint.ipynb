```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/stock_day.csv")
df.head()
```

```python
# 1. pd.qcut：指定区间的数量，pd.qcut会自动计算切割点
qcut = pd.qcut(df['p_change'], 6)
qcut.value_counts()
```

```python
# 2. pd.cut：手动指定分割的区间
cut = pd.cut(df['p_change'], [-15, -10, -5, 0, 5, 10, 15])
cut.value_counts()
```
