# 二、pd.merge

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 5, 5],
                   'xx': ['11', '22', '33', '44']})
                    
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8],
                   'xx': ['44', '202', '303', '404']})

print(df1)
print()
print(df2)
```

## 1. 内连接：

```python
# 内连接：默认会寻找两个dataframe共同的列的值，看是否相同，如果相同就会认为是可以连接的，就会保留
# 也可以通过on参数，手动指定连接的字段
# pd.merge(df1, df2, how="inner")
pd.merge(df1, df2, how="inner", on="value")
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
                    
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})

print(df1)
print()
print(df2)
```

```python
# 也可以使用left_on和right_on来指定左右dataframe连接的字段
pd.merge(df1, df2, left_on="lkey", right_on="rkey")
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})

print(left)
print()
print(right)
```

```python
pd.merge(left, right, on=['key1', 'key2'])
```

## 2. 外连接：

```python
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

## 3. 左连接：

```python
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

## 4. 右连接：

```python
pd.merge(left, right, how='right', on=['key1', 'key2'])
```
