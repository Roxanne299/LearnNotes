```python
import numpy as np
```

```python
a = np.arange(10)
print(a)
print(a.dtype)
```

```python
b = np.array([1,2,3,4,5,],dtype='i1')
print(b)
print(b.dtype)
```

```python
c = np.array([1,2,3,4,5],dtype=np.float16)
print(c)
print(c.dtype)
```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
d = np.array([Person('知了',18),Person('课堂',20)])
print(d)
print(d.dtype)
```

```python
f = np.array(['a','b'],dtype='S')
print(f)
print(f.dtype )
```

```python
e = np.array(['a','c'],dtype='U')
print(e)
print(e.dtype)
```

```python
uf = f.astype("U")
print(uf)
print(uf.dtype)
print(f.dtype)
```

    * Numpy本身是基于C语言编写，C语言中本身就是有很多数据类型，所以直接引用过来了。
    * Numpy为了考虑到处理海量数据的性能，针对不同的数据给不同的数据类型，来节省内存空间，所以有不同的数据类型。
