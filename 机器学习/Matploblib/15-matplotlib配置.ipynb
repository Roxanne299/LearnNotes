```python
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
plt.plot(range(0,10))
plt.plot(range(0,10,2))
```

```python
from matplotlib import cycler
plt.rcParams['axes.prop_cycle'] = cycler('color', ['k', 'y'])
plt.rcParams['font.sans-serif'] = ['STHupo']
plt.rcParams['font.size'] = 20
plt.plot(range(0,10))
plt.plot(range(0,10,2))
plt.title("测试")
```

```python
mpl.get_configdir()
```

# 笔记:
1. 动态设置配置:通过plt.rcParams来设置.
2. 配置文件配置:把配置项放到[名称].pltstyle文件中,然后把这个文件放在`matplotlib.get_configdir()/stylelib`下,以后在代码中就通过`plt.style.use("名称")`就可以使用了.
