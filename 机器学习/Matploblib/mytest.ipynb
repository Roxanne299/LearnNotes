```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager
```

```python
y = [np.random.randint(1,10) for x in range(10)]
x = range(2,12)
plt.plot(x,y)
```

```python
mydata = {"x":range(2,12),"y":[np.random.randint(1,10) for x in range(10)]}
plt.plot("x","y","^",data=mydata,color='red')
```

```python
plt.plot(x,y,linewidth=10)
```

```python
line2D = plt.plot(x,y)
line = line2D[0]
line.set_linewidth(10)
```

```python
line2D = plt.plot(x,y,range(10),range(10))
plt.setp(line2D[1],linewidth=10)
```

```python
font = font_manager.FontProperties(fname=r"C:\Windows\Fonts\msyh.ttc",size=20)
plt.figure(figsize=(15,10))
plt.plot(x,y)
plt.title("折线图",fontproperties=font)
plt.xticks(range(12),["%d天"%x for x in range(12)],fontproperties=font)
plt.xlabel("天数",fontproperties=font)
plt.ylabel("票房（单位：w）",fontproperties=font)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager
highest = [26,21,26,26,22,20,17,19,22,28,30,28,24,28,25,26,25,26,25,23,24,30,32,31,30,27,26,27,29,25,25]
lowest =  [17,13,17,18,18,17,14,15,16,18,19,20,18,18,20,20,20,20,20,16,17,19,21,24,24,23,20,18,19,18,19]
font = font_manager.FontProperties(fname=R"C:\Windows\Fonts\msyh.ttc")
x = np.arange(1,32,1)
plt.figure(figsize=(20,9),dpi=200)
plt.plot(x,highest,color='red',marker='o')
plt.xticks(range(1,32))
plt.xlabel("天",fontproperties=font)
plt.ylabel("气温（摄氏度）",fontproperties=font)
for index,value in enumerate(highest):
    plt.annotate("%d"%value,xy=(index,value),xytext=(index+1,value+0.1))
plt.plot(x,lowest,color='blue',marker='o')
for index,value in enumerate(lowest):
    plt.annotate("%d"%value,xy=(index,value),xytext=(index+1,value+0.1))
plt.grid(True)
```
